```python
import camelot
import pandas as pd

PDF_PATH = "/content/2023 age.pdf"
OUTPUT_EXCEL = "2023 age.xlsx"
```

```python
print("📄 Reading tables from PDF...")

tables = camelot.read_pdf(
    PDF_PATH,
    pages="all",
    flavor="stream")

print(f"✅ Tables found: {tables.n}")
```

```python
all_dfs = []

for i, table in enumerate(tables):
    df = table.df
    df["source_table"] = i + 1
    all_dfs.append(df)

final_df = pd.concat(all_dfs, ignore_index=True)
```

```python
final_df.to_excel(OUTPUT_EXCEL, index=False)

print(f"📊 Excel saved as: {OUTPUT_EXCEL}")
```
